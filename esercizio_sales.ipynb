```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
###
##  X label name = Month Number
##  Y label name = Total profit
##
df =pd.read_csv('https://frenzy86.s3.eu-west-2.amazonaws.com/fav/company_sales_data.csv')
```

```python
import calendar

for month_idx in range(1, 13):
    print (calendar.month_name[month_idx])
```

```python
month_word = {
1:'January',
2:'February',
3:'March',
4:'April',
5:'May',
6:'June',
7:'July',
8:'August',
9:'September',
10:'October',
11:'November',
12:'December'
}
```

```python
df['month_number'] = df['month_number'].map(month_word)

```

```python
df = df.rename(columns={'month_number':'month'})
df
```

```python
item = df.columns[1:-2].tolist()
```

```python
plt.figure(figsize=(20,10))
ax = plt.gca()
for i in item:
    df.plot(kind="line", x="month", y= i, ax = ax)
```
